Linear Regression and Non-Linear Regression Relationship


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('linear_data.csv')
df.head()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df[['x']], df['y'])
```

```python
# Linear
plt.figure(figsize=(10,5))
plt.scatter(1,2,1)

plt.scatter(df.x, df.y, color="blue", label="Data with Linear Relationship")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Linear Relationship")
plt.legend()

# Non-Linear
plt.subplot(1,2,2)
plt.scatter(df2.x, df2.y, color='green', label="Data with Non-linear Relationship")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non-Linear Relationship')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
correlation_linear = np.corrcoef(df.x, df.y)
print("Correlation coefficient for linear relationship:\n",correlation_linear)
```

```python
correlation_nonlinear = np.corrcoef(df2.x, df2.y)[0,1]
print("Correlation coefficient for non-linear relationship:", correlation_nonlinear)
```

```python
from sklearn.linear_model import LinearRegression
df.head()
```

```python
reg1 = LinearRegression()
```

```python
reg1.fit(df[['x']], df[['y']])
```

```python
reg1.score(df[['x']], df[['y']])
```

```python
x= df[['x']]
y = df['y']
```

```python
y.shape
```

```python
reg1.score(x,y)
```

```python
from sklearn.metrics import r2_score
r2_score(y, reg1.predict(x))
```

```python
plt.scatter(df[['x']], df.y, color="blue", label="Data with Linear Relationship")
plt.plot(df[['x']], reg1.predict(df[['x']]), color="red", label="Best fit line")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Relationship')
plt.legend()
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(df2[['x']])
```

```python
x_poly.shape
```

```python
x_poly
```

```python
pd.DataFrame(x_poly).head()
```

```python
df2.x.head()
```

Polynomial Regression = Poly Features + Linear Regression

```python
reg_poly = LinearRegression()
```

```python
reg_poly.fit(x_poly, df2.y)
```

```python
reg_poly.score(x_poly,df2.y)
```

```python
plt.figure(figsize=(9,6))
plt.scatter(df2.x, df2.y, color="green", label="Data with Non-linear Relationship")
plt.plot(df2.x, 2*np.sin(df2.x), color="red", label="True Non-linear Relationship")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Non Linear Relationship")

plt.legend()
plt.tight_layout()
plt.show()
```
